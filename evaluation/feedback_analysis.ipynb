```python
answer = ['Strongly disagree', 'Disagree', 'Neither agree nor disagree', 'Agree', 'Strongly agree']
prev = [0, 2, 4, 7, 3]
curr = [0, 0, 9, 9, 16]
prev = [i/sum(prev) for i in prev]
curr = [i/sum(curr) for i in curr]
```

```python
print(prev)
print(curr)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x_axis = np.arange(len(answer))


plt.bar(x_axis -0.2, prev, width=0.4, label = 'Pervious Version')
plt.bar(x_axis +0.2, curr, width=0.4, label = 'Current Version')

plt.xticks(x_axis, answer,     rotation=30, 
    horizontalalignment='right',
    fontweight='light', )
plt.ylabel('Percentage')
plt.title('The chatbot was good at guessing my emotion')

plt.legend()
plt.show()
```
