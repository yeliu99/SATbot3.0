```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_data = pd.read_csv('/Users/yeliu/IC/Individual_Project/code/sat3.0/data/emotions-dataset-for-nlp/unprocessed_dataset_emotions/train.txt', names=['text', 'emotion'], sep=';')
val_data = pd.read_csv('/Users/yeliu/IC/Individual_Project/code/sat3.0/data/emotions-dataset-for-nlp/unprocessed_dataset_emotions/val.txt', names=['text', 'emotion'], sep=';')
test_data = pd.read_csv('/Users/yeliu/IC/Individual_Project/code/sat3.0/data/emotions-dataset-for-nlp/unprocessed_dataset_emotions/test.txt', names=['text', 'emotion'], sep=';')
```

```python
data = {'Training Data': train_data, 'Validation Data': val_data, 'Test Data': test_data}

bar, ax = plt.subplots(1,3, figsize=(30, 10))
for index, temp in enumerate(data):
    sns.countplot(ax = ax[index],x = 'emotion', data = data[temp])
    ax[index].set_title(temp+' Class Frequency', size=14)
    ax[index].set_ylabel('Frequency', size=14)
    #ax[index].set_xlabel(temp+' Class', size=14)
```
