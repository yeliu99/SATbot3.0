```python
from parrot import Parrot
parrot = Parrot()
```

```python
import pandas as pd
user_rewrite = pd.read_csv("drive/MyDrive/user_rewritings.csv", names=['utterance'])
```

```python
input_phrases = user_rewrite['utterance'].tolist()
```

```python
phrases = ["Can you recommed some upscale restaurants in Newyork?",
           "What are the famous places we should not miss in Russia?"
]

for phrase in phrases:
  print("-"*100)
  print("Input_phrase: ", phrase)
  print("-"*100)
  para_phrases = parrot.augment(input_phrase=phrase)
  for para_phrase in para_phrases:
   print(para_phrase)
```

```python
paraphrase = []

for phrase in input_phrases:
  print("Input_phrase: ", phrase)
  para_phrases = parrot.augment(input_phrase=phrase)
  if para_phrases != None:
    paraphrase.append(para_phrases[0][0])
    for para_phrase in para_phrases:
      print(para_phrase)
  else:
    paraphrase.append(phrase)
```

```python
user_rewrite['paraphrased'] = paraphrase
user_rewrite.to_csv("drive/MyDrive/parrot_paraphrased.csv", index=False)
```
